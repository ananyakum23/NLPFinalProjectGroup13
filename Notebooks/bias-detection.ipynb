```python
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("d4data/bias-detection-model")
model = TFAutoModelForSequenceClassification.from_pretrained("d4data/bias-detection-model")
```

```python
from transformers import pipeline
import pprint

classifier = pipeline('text-classification', model=model, tokenizer=tokenizer)
```

```python
# Specify file path for output
output_file_path = "/content/drive/MyDrive/NLP/bias-detection-model/twitter_pos_results.csv"

# Open a new file to write the results
with open(output_file_path, "w") as output_file:
    output_file.write("Line Number, Positive Text, Label, Score\n")  # Write header
    j = 0
    for line_number, line in enumerate(data, start=1):

        line = line.strip()
        try:
            review = line.split(",")[0]  # Assuming the review is the second column
        except:
          continue

        # Perform sentiment analysis
        result = classifier(review)
        print(line_number)

        score = result[0]['score']
        label = result[0]['label']




        # Write line number, review, label, and score to the output file
        output_file.write(f"{line_number}, {review}, {label}, {score}\n")


print("Results saved to 'twitter_pos_results.csv'")
```
